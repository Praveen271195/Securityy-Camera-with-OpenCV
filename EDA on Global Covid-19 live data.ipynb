```python
import pandas as pd
```

```python
recovered = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")
confirmed = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
```

## Exploratory Data Analysis

```python
# Head of the "confimred" dataset
confirmed.head()
```

```python
# No. of rows and columns
confirmed.shape
```

```python
# Names of the columns in dataset
confirmed.columns
```

```python
# Reshaping the "confirmed" dataset
confirmed = pd.melt(confirmed,id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],var_name="Date",value_name="Confirmed_Cases")
confirmed.tail()
```

```python
# Checking the dataset with the data available on other sources
confirmed[confirmed["Country/Region"]=="India"].tail()
```

```python
# Head of the "recovered" dataset
recovered.head()
```

```python
# Reshaping the recovered dataset
recovered = pd.melt(recovered,id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],var_name="Date",value_name="Recovered_cases")
```

```python
# Head of the "deaths" dataset
deaths.head()
```

```python
# Reshaping the "deaths" dataset
deaths = pd.melt(deaths,id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],var_name="Date",value_name="Deaths")
```

```python
# Checking the deaths dataset
deaths.tail()
```

```python
# Merging the above two datasets - "Confirmed cases and Deaths"
merge1 = pd.merge(right = deaths, how = 'left',on = ['Province/State', 'Country/Region', 'Lat', 'Long', 'Date'],left = confirmed)
merge1.tail()
```

```python
# Merging the above merged data with the "Recovered" dataset
merge2 = pd.merge(left=merge1,right=recovered,on=['Province/State', 'Country/Region', 'Lat', 'Long', 'Date'])
merge2.tail()
```

```python
# Renaming the above merge2 as covid_data
covid_data = merge2
```

```python
# Date is in String format
covid_data.info()
```

```python
# Converting the string dates into data time format
covid_data["Date"] = pd.to_datetime(covid_data["Date"])
covid_data.head()
```

```python
# Checking the null values
covid_data.isnull().sum()
```

```python
# Adding the active cases information
covid_data["Active_cases"] = covid_data["Confirmed_Cases"]-covid_data["Deaths"]-covid_data["Recovered_cases"]
```

```python
# Checking Covid data 
covid_data[covid_data["Country/Region"]=="India"].tail() #we are having a dataframe that is cumilative and we find the difference between the values for finding the increase or decrease
```

```python
# Latest date in required format
covid_data.Date.iloc[-1].strftime("%dth %B, %Y")
```

```python
# Sum of all required columns with respect to the date column
cum_global_cases = covid_data.groupby(['Date'])[['Confirmed_Cases', 'Deaths', 'Recovered_cases', 'Active_cases']].sum().reset_index()
cum_global_cases.tail()
```

```python
# Latest total Confirmed Cases 
cum_global_cases["Confirmed_Cases"].iloc[-1]
```

```python
# Latest total Recovered cases 
cum_global_cases["Recovered_cases"].iloc[-1]
```

```python
# Latest total Active cases cases 
cum_global_cases["Active_cases"].iloc[-1]
```

```python
# Percent change between the previous and latest Global Confirmed Cases 
round(((cum_global_cases["Confirmed_Cases"].iloc[-1]-cum_global_cases["Confirmed_Cases"].iloc[-2])/cum_global_cases["Confirmed_Cases"].iloc[-2])*100,4)
```

```python
# List of Countries for drop down 
covid_data["Country/Region"].unique()
```

```python
# Grouped data based on date and country
grouped_by_date_and_country = covid_data.groupby(['Date','Country/Region'])[['Confirmed_Cases', 'Deaths', 'Recovered_cases', 'Active_cases']].sum().reset_index()
```

```python
# Difference between the latest confirmed cases and previous day's confirmed cases 
value_confirmed = grouped_by_date_and_country[grouped_by_date_and_country["Country/Region"]=="India"]["Confirmed_Cases"].iloc[-1]-grouped_by_date_and_country[grouped_by_date_and_country["Country/Region"]=="India"]["Confirmed_Cases"].iloc[-2]
value_confirmed
```

```python
# Difference between the Yesterday's confirmed cases and it's previous day's confirmed cases
grouped_by_date_and_country[grouped_by_date_and_country["Country/Region"] == "India"]["Confirmed_Cases"].iloc[-2]-grouped_by_date_and_country[grouped_by_date_and_country["Country/Region"] == "India"]["Confirmed_Cases"].iloc[-3]
```

```python
# Dataframe that shows that shows daily confirmed cases for selected country
grouped_by_date_and_country = covid_data.groupby(['Date', 'Country/Region'])[['Confirmed_Cases', 'Deaths', 'Recovered_cases', 'Active_cases']].sum().reset_index()
daily_confirmed_cases1 = grouped_by_date_and_country[grouped_by_date_and_country["Country/Region"]=="India"][["Country/Region","Date","Confirmed_Cases"]].reset_index()
daily_confirmed_cases1
```

```python
# Adding a new column for calculating the daily confirmed cases. 
daily_confirmed_cases1["Daily_confirmed_cases"] = daily_confirmed_cases1["Confirmed_Cases"]-daily_confirmed_cases1["Confirmed_Cases"].shift(1)
```

```python
# Rolling average
daily_confirmed_cases1["Rolling_avg"] = daily_confirmed_cases1["Daily_confirmed_cases"].rolling(window=7).mean()
```
